```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# only approved cases
data20 = pd.read_excel("data/PERM_Disclosure_Data_FY20.xlsx")
data21 = pd.read_excel("data/PERM_Disclosure_Data_FY21.xlsx")
data22 = pd.read_excel("data/PERM_Disclosure_Data_FY22.xlsx")
data23 = pd.read_excel("data/PERM_Disclosure_Data_FY23.xlsx")
data24 = pd.read_excel("data/PERM_Disclosure_Data_FY24.xlsx")

data24.head(5)
```

```python
data20 = data20[['CASE_NUMBER', 'CASE_STATUS', 'EMPLOYER_NAME', 'NAICS_CODE', 'EMPLOYER_NUM_EMPLOYEES', 'COUNTRY_OF_CITIZENSHIP', 'PW_SKILL_LEVEL', 'PW_WAGE']]
data21 = data21[['CASE_NUMBER', 'CASE_STATUS', 'EMPLOYER_NAME', 'NAICS_CODE', 'EMPLOYER_NUM_EMPLOYEES', 'COUNTRY_OF_CITIZENSHIP', 'PW_SKILL_LEVEL', 'PW_WAGE']]
data22 = data22[['CASE_NUMBER', 'CASE_STATUS', 'EMPLOYER_NAME', 'NAICS_CODE', 'EMPLOYER_NUM_EMPLOYEES', 'COUNTRY_OF_CITIZENSHIP', 'PW_SKILL_LEVEL', 'PW_WAGE']]
data23 = data23[['CASE_NUMBER', 'CASE_STATUS', 'EMPLOYER_NAME', 'NAICS_CODE', 'EMPLOYER_NUM_EMPLOYEES', 'COUNTRY_OF_CITIZENSHIP', 'PW_SKILL_LEVEL', 'PW_WAGE']]
data24 = data24[['CASE_NUMBER', 'CASE_STATUS', 'EMPLOYER_NAME', 'NAICS_CODE', 'EMPLOYER_NUM_EMPLOYEES', 'COUNTRY_OF_CITIZENSHIP', 'PW_SKILL_LEVEL', 'PW_WAGE']]
data24
```

```python
# filtering certified and certified expired (as they both mean certified)
mask = (data20['CASE_STATUS'] == 'Certified') | (data20['CASE_STATUS'] == 'Certified-Expired') & (data20['EMPLOYER_NUM_EMPLOYEES'] > 10000)
data20 = data20[mask]

mask = (data21['CASE_STATUS'] == 'Certified') | (data21['CASE_STATUS'] == 'Certified-Expired') & (data20['EMPLOYER_NUM_EMPLOYEES'] > 10000)
data21 = data21[mask]

mask = (data22['CASE_STATUS'] == 'Certified') | (data22['CASE_STATUS'] == 'Certified-Expired') & (data20['EMPLOYER_NUM_EMPLOYEES'] > 10000)
data24 = data22[mask]

mask = (data23['CASE_STATUS'] == 'Certified') | (data23['CASE_STATUS'] == 'Certified-Expired') & (data20['EMPLOYER_NUM_EMPLOYEES'] > 10000)
data24 = data23[mask]

mask = (data24['CASE_STATUS'] == 'Certified') | (data24['CASE_STATUS'] == 'Certified-Expired') & (data20['EMPLOYER_NUM_EMPLOYEES'] > 10000)
data24 = data24[mask]
```
